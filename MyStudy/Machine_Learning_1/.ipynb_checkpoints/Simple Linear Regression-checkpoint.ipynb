```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('slr_excel_demonstration_1.csv')
df
```

```python
import numpy as np
import seaborn as sns

sns.set_style("whitegrid")


sns.regplot(x=df['Marketing Budget (X) (In lakhs)'], y=df['Actual Sales(Y) (In crores)'])
plt.show()
```

```python
# Use python sklearn module for linear regression predictions
from sklearn.linear_model import LinearRegression

X = df.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = df.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column

linear_regressor = LinearRegression()  # create object for the class
linear_regressor.fit(X, Y)  # perform linear regression
df['Predicted Sales (Y-pred)'] = linear_regressor.predict(X)  # make predictions
df
```

```python
#For retrieving the slope:
print("For line y = b0 + b1x, where b1 is slope.")
print(linear_regressor.coef_)
```

```python
#To retrieve the intercept:
print("For line y = b0 + b1x, where b0 is intercept.")
print(linear_regressor.intercept_)
```

```python
df['Residual Squares'] = (df['Actual Sales(Y) (In crores)'] - df['Predicted Sales (Y-pred)'])**2
df
```

```python
RSS = df['Residual Squares'].sum()
print("Residual Sum of Squares (RSS) is {}".format(RSS))
```

```python
Average_Sales = df['Actual Sales(Y) (In crores)'].mean()
Average_Sales
```

```python
df['Sum of Squares(R^2)'] = (df['Actual Sales(Y) (In crores)'] - Average_Sales)**2
df
```
