```python
#rounds2.columns
rounds2 = rounds2[np.isfinite(rounds2['raised_amount_usd'])]
rounds2.head()
```

```python
rounds2_temp = rounds2.copy(deep=True)
#rounds2_temp.rename(columns = {'company_permalink':'permalink'}, inplace=True)
```

```python
rounds2_temp.head()
```

```python
def str_to_lower(string):
    return string.lower()

rounds2_temp['permalink_lower'] = rounds2_temp.company_permalink.apply(str_to_lower)
```

```python
rounds2_temp.head()
```

```python
companies_temp = companies.copy(deep=True)
companies_temp['permalink_lower'] = companies_temp.permalink.apply(str_to_lower)
companies_temp.head()
```

```python
master_frame = pd.merge(rounds2_temp, companies_temp, how='inner', on='permalink_lower')
#master_frame = pd.merge(rounds2_temp, companies_temp, how='inner', left_on='round2.company_permalink', right_on='companies.permalink')
```

```python
master_frame.head(100)
```

```python
master_frame.shape
```

```python
fund_type = ['venture', 'angel', 'seed', 'private_equity']

fundtype_to_invest = master_frame.loc[master_frame['funding_round_type'].isin(fund_type)]

values = {'raised_amount_usd': 0 }
fundtype_to_invest.fillna(value=values, inplace=True)
fundtype_to_invest
```

```python
df = fundtype_to_invest.loc[:, ['funding_round_type','raised_amount_usd']]
df_by_funding_type = df.groupby('funding_round_type')
df_by_funding_type.raised_amount_usd.mean()
```

```python
#value_in_range = 5000000 <= fundtype_to_invest.raised_amount_usd
#fund_value_in_range = fundtype_to_invest[value_in_range]
#value_in_range = 15000000 >= fund_value_in_range.raised_amount_usd
#fund_value_in_range = fund_value_in_range[value_in_range]
#fund_value_in_range
fund_value_in_range = fundtype_to_invest.loc[(fundtype_to_invest.raised_amount_usd >= 5000000) & \
                                             (fundtype_to_invest.raised_amount_usd <= 15000000),:]
fund_value_in_range
```

```python
df = fund_value_in_range.groupby('funding_round_type').count()
df.loc[['venture','angel','seed','private_equity'],['permalink_lower']]
```

### Thus venture funding is the most preffered investment type for Spark Funds

```python
preffered_fund = fund_value_in_range.loc[fund_value_in_range['funding_round_type'] == 'venture']
preffered_fund
```

# Checkpoint 3 - country analysis.
#### Spark Funds wants to see the top nine countries which have received the highest total funding 

```python
df = preffered_fund.loc[:, ['country_code','raised_amount_usd']]
df_by_country = df.groupby('country_code')
ds = df_by_country.raised_amount_usd.sum().sort_values(ascending=False)
ds[0:9]

#preffered_fund.raised_amount_usd.sort_values(ascending=False)

#preffered_fund.sort_values('raised_amount_usd', ascending=False)
```

```python
top9 = preffered_fund.loc[preffered_fund['country_code'].isin(ds[0:9].index),:]
top9
```

```python
top3_eng_country = ['USA', 'GBR', 'CAN']
top3 = top9.loc[top9['country_code'].isin(top3_eng_country),:]
```

```python
top3
```
