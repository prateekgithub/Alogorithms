```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Plotting two 1-D numpy arrays
x = np.linspace(5, 100, 100) # 100 numbers between 5 to 100
y = np.linspace(10, 1000, 100) # 100 numbers between 10 to 1000

# create a plot obj
plt.plot(x,y)

# need to call plt.show() explicitly to display the plot
plt.show(block=False)
plt.show()
```

```python
# can also work with lists, though it converts lists to np arrays internally
plt.plot([1,4,6,8],[3,8,3,5])
plt.show()
```

```python
# Axis labels and title

plt.plot(x,y)

# x and y labels and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

# Define the range of labels of the axis
# Arguments: plt.axis(xmin, xmax, ymin, ymax)
plt.xlim([20, 80])
plt.ylim([200, 800])
plt.show()
```

```python
# Change the color and line type

# initializing x and y arrays
x = np.linspace(0, 10, 20)
y = x*2

# Color blue, line type +
plt.plot(x, y, 'b+')

# x and y labels and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's Law")

plt.show()
```

```python
# Plotting multiple lines on the same plot

x = np.linspace(0, 5, 10)
y = np.linspace(3, 6, 10)

# plot three curves: y, y**2 and y**3 with different line type
plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')
plt.show()
```

Figure and subplots - using which you can create multiple subplots inside the same figure.

```python
x = np.linspace(1, 10 ,100)
y = np.log(x)

# initiate a new figure explicitly.
plt.figure(1)

# Create subplots with 1 row and 2 columns

# Create first subplot in figure 1
plt.subplot(1, 3, 1)
plt.title("y = log(x)")
plt.plot(x,y)

# create a second subplot in figure-1
plt.subplot(1, 3, 2)
plt.title("y = log(x)**2")
plt.plot(x,y**2)

# create a third subplot in figure-1
plt.subplot(1, 3, 3)
plt.title("y = log(x)**3")
plt.plot(x,y**3)

plt.show()
```

```python
# initiate a new figure explicitly.
plt.figure(2)

# Create subplots with 1 row and 2 columns

# Create first subplot in figure 1
plt.subplot(2, 2, 1)
plt.title("y = log(x)")
plt.plot(x,y)

# create a second subplot in figure-1
plt.subplot(2, 2, 2)
plt.title("y = log(x)**2")
plt.plot(x,y**2)

# create a third subplot in figure-1
plt.subplot(2, 2, 3)
plt.title("y = log(x)**3")
plt.plot(x,y**3)

plt.show()
```

## Univariate Plots

```python
import seaborn as sns

# set a seaborn style of your taste
sns.set_style("whitegrid")

# data
df = pd.read_csv("./global_sales_data/market_fact.csv")
df.head()
```

### Histogram and Density Plot

```python
# simple density plot
sns.distplot(df.Shipping_Cost)
```

```python
# rug = True
# plotting only a few points since rung takes time.

sns.distplot(df['Shipping_Cost'][:100], bins=20)
plt.show()
```

```python
# Subplots

# Subplot 1
plt.subplot(2, 2, 1)
plt.title('Sales')
sns.distplot(df.Sales)

# Subplot 2
plt.subplot(2, 2, 2)
plt.title('Profit')
sns.distplot(df.Profit)

# Subplot 3
plt.subplot(2, 2, 3)
sns.distplot(df.Order_Quantity)

# Subplot 4
plt.subplot(2, 2, 4)
sns.distplot(df.Shipping_Cost)

plt.show()
```

### Boxplots

```python
# Boxplot

sns.boxplot(df.Order_Quantity)
plt.title('Order Quantity')

plt.show()
```

```python
# Boxplot Y oriented

sns.boxplot(y=df.Order_Quantity)
plt.title('Order Quantity')

plt.show()
```

```python
sns.distplot(df['Sales'][:1000], rug=True)
```

## Bivariate Plots

### Simple Scatter plot

```python
df.plot.scatter(x='Sales', y='Profit')
```

```python
# Joint plot of profit and loss

sns.jointplot('Sales','Profit', df)
# Same as sns.joinplot(df['Sales'],df['Profit'])
```

```python
df2 = df[(df.Profit < 10000) & (df.Sales < 40000)]
sns.jointplot('Sales','Profit', df2)

plt.show()
```
