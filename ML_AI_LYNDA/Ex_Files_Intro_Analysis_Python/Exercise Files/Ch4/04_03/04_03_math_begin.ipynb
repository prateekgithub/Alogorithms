```python
import numpy as np
import matplotlib.pyplot as pp

%matplotlib inline
```

```python
x = np.linspace(0, 10, 40)
```

```python
x
```

```python
sinx = np.sin(x)
```

```python
pp.plot(x,sinx)
```

```python
cosx = np.cos(x)
pp.plot(x,cosx)
```

```python
pp.plot(x,sinx,'x')
pp.plot(x,cosx,'o')
```

```python
pp.plot(x,sinx)
pp.plot(x,cosx)
pp.legend(['sinx','cosx'])
```
