```python
wordclean = [word.strip().lower() for word in open('words','r')]
```

```python
wordclean[:10]
```

```python
wordclean = sorted(list(set([word.strip().lower() for word in open('words','r')])))
```

```python
wordclean[:10]
```

```python
def signature(myword):
    return ''.join(sorted(myword))
```

```python
signature('dictionary')
```

```python
def anagram(myword):
    return [ word for word in wordclean if signature(myword) == signature(word) ]
```

```python
anagram('dictionary')
```

```python
import collections
```

```python
words_bysig = {}
words_bysig = collections.defaultdict(list)
for word in wordclean:
    words_bysig[signature(word)].append(word)
```

```python
words_bysig
```

```python
def anagram_fast(myword):
    return words_bysig[signature(myword)]
anagram_fast('dictionary')
```

```python
anagram_all = { word: anagram_fast(word) for word in wordlist if len(anagram_fast(word)) > 1}
```

```python
len(anagram_all)
```
